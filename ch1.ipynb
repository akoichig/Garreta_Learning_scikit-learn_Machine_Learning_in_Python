```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
iris = datasets.load_iris()
X_iris, y_iris = iris.data, iris.target
print X_iris.shape, y_iris.shape
```

```python
print X_iris[0], y_iris[0]
```

```python
from sklearn.cross_validation import train_test_split
from sklearn import preprocessing
# Get dataset iwth only the first two attributes
X, y = X_iris[:,:2], y_iris
# Split the dataset into a training and a testing set
# Test set will be the 25% taken randomly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)
```

```python
print X_train.shape, y_train.shape
```

```python
# Standardize the features
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
colors = ['red', 'greenyellow', 'blue']
for i in xrange(len(colors)):
    xs = X_train[:, 0][y_train == i]
    ys = X_train[:, 1][y_train == i]
    plt.scatter(xs, ys, c=colors[i], s=60)
plt.legend(iris.target_names)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
```

```python
# Stochastic Gradient Descent
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier()
clf.fit(X_train, y_train)
```

```python
print clf.coef_
```

```python
print clf.intercept_
```

```python
# Draw three decision boundaries
x_min, x_max = X_train[:, 0].min() - .5, X_train[:, 0].max()+.5
y_min, y_max = X_train[:, 1].min() - .5, X_train[:, 1].max()+.5
xs = np.arange(x_min, x_max, 0.5)
```

```python
fig, axes = plt.subplots(1,3)
fig.set_size_inches(20,12)
for i in [0,1,2]:
    axes[i].set_aspect('equal')
    axes[i].set_title('Class '+ str(i) + ' versus the rest')
    axes[i].set_xlabel('Sepal length')
    axes[i].set_ylabel('Sepal width')
    axes[i].set_xlim(x_min, x_max)
    axes[i].set_ylim(y_min, y_max)
    plt.sca(axes[i])
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.prism, s=60)
    ys = (-clf.intercept_[i] - xs * clf.coef_[i, 0]) / clf.coef_[i, 1]
    plt.plot(xs, ys, hold=True)
```

```python
print clf.predict(scaler.transform([[4.7, 3.1]]))
```

```python
print clf.decision_function(scaler.transform([[4.7, 3.1]]))
```

```python
# Evaluating Results
from sklearn import metrics
y_train_pred = clf.predict(X_train)
print metrics.accuracy_score(y_train, y_train_pred) # the % of training set instances correclty classified
```

```python
# Evaluate classifier on reserved test set
y_pred = clf.predict(X_test)
print metrics.accuracy_score(y_test, y_pred)
```

```python
# Accuracy (used above) is a good performance measure when the number of instances of each class is similar, 
# but not for skewed distributions. (TP+TN) / m

# PRECISION: proportion of instances predicted as positive that were correctly evaluted
# TP / (TP + FP)

# RECALL: proportion of positive instances that were correctly evaluated
# TP / (TP + FN)

# F1-SCORE: harmonic mean of precision and recall
# 2 * Precision * Recall / (Precision + Recall)

print metrics.classification_report(y_test, y_pred, target_names=iris.target_names)
```

```python
# Confusion matrix
# (i, j) entry is number of class instances i predicted to be in class j
print metrics.confusion_matrix(y_test, y_pred)
```

```python
# Cross validation
from sklearn.cross_validation import cross_val_score, KFold
from sklearn.pipeline import Pipeline
# create a composite estimator made by a pipeline of the standardization and the linear model
clf = Pipeline([
        ('scaler', preprocessing.StandardScaler()),
        ('linear_model', SGDClassifier())
    ])
```

```python
# create a k-fold cross validation iterator of k=5 folds
cv = KFold(X.shape[0], 5, shuffle=True, random_state=33)
```

```python
# by default the score used is the one returned by score method of the estimator (accuracy)
scores = cross_val_score(clf, X, y, cv=cv)
print scores
```

```python
from scipy.stats import sem
def mean_score(scores):
    return ("Mean score: {0:.3f} (+/-{1:.3f})").format(np.mean(scores), sem(scores))
print mean_score(scores)
```
